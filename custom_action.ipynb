```python
datasource_name = "input_datasource"
datasource = context.datasources.get(datasource_name)
if datasource is None:
    datasource  = context.sources.add_pandas(name=datasource_name)
```

```python
df = pd.read_csv("mock_data.csv")
```

```python
data_asset_name = "test_dataframe"

data_asset = None
for asset in datasource.assets:
    if asset.name == data_asset_name:
        data_asset = asset
        break
if data_asset is None:
    data_asset = datasource.add_dataframe_asset(name=data_asset_name)
```

```python
batch_request = data_asset.build_batch_request(dataframe=df)
```

```python
validator = context.get_validator(
    batch_request=batch_request, create_expectation_suite_with_name="new_expectation_suite"
)
```

```python
df
```

```python
validator.expect_column_value_lengths_to_equal(column="Name",value=13)
```

```python
validator.expect_column_value_lengths_to_equal(column="AccountBalance",value=6)
```

```python
validator.expect_column_value_lengths_to_equal(column="DateOfBirth",value=10)
```

```python
validator.save_expectation_suite(discard_failed_expectations=False)
```

```python
checkpoint = context.add_or_update_checkpoint(
    name="test_checkpoint",
    validator=validator,
    runtime_configuration={
        "result_format": {
            "result_format": "COMPLETE"
        },
    },
    action_list=[
        {
            "name": "my_custom_action",
            "action": {"class_name": "MyCustomAction",
                      "module_name": "plugins.my_custom_action",},
        }
    ]
)
```

```python
checkpoint_result = checkpoint.run()
checkpoint_result
```
